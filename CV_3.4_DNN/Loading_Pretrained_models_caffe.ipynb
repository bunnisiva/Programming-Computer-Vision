# Input an image, preprocess it and convert it into tensor 

```python
image = cv2.imread('../data/volleyball05.jpg')
tensor = cv2.dnn.blobFromImage(image,1.0,(224,224),(104,117,123),False,False)
```

```python
# Load a caffe model 
net_caffe = cv2.dnn.readNetFromCaffe('../data/bvlc_googlenet.prototxt','../data/bvlc_googlenet.caffemodel')
```

```python
# To view the architecture please look at the prototxt file . 
```

```python
net_caffe.setInput(tensor);
prob = net_caffe.forward();
```

```python
prob[0][0:5]
```

```python
net_caffe.setInput(tensor,"data");
prob = net_caffe.forward('prob');
```
